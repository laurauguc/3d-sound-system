Additional lines:

* line 3: (SALA_WIDTH, TMLp_y, 3.3) --> (SALA_WIDTH, SALA_DEPTH / 2, 3.3). Name = f"Distance from center = {side_shift}"

* line 4: (SALA_WIDTH, TMLR_y, 3.3) --> (SALA_WIDTH, SALA_DEPTH / 2, 3.3). Name = f"Distance from center = {side_shift}"


```python
from dataclasses import dataclass
import numpy as np
import plotly.graph_objects as go


# =========================================================
# CONFIGURATION
# =========================================================

SALA_WIDTH = 4.25
SALA_DEPTH = 4.13

STAGE_WIDTH = 3.70
STAGE_DEPTH = SALA_DEPTH
STAGE_HEIGHT = .6 # double check

# Ceiling disk coordinates
side_shift = 1.65
forward_shift = 1.3

TMLp_y = round(SALA_DEPTH / 2 - side_shift,2)
TMRp_y = round(SALA_DEPTH / 2 + side_shift,2)
TMp_x = round(SALA_WIDTH - forward_shift,2)


@dataclass
class BoxObject:
    name: str
    coords: list[tuple[float, float]]
    height: float
    color: str
    base_z: float = 0.0


# =========================================================
# GEOMETRY HELPERS
# =========================================================

def rectangle(x: float, y: float, width: float, depth: float):
    """Return rectangle corner coordinates."""
    return [
        (x, y),
        (x + width, y),
        (x + width, y + depth),
        (x, y + depth),
    ]


def centered_against_wall(
    room_width: float,
    room_depth: float,
    obj_width: float,
    obj_depth: float,
):
    """Place object centered on the right wall."""
    x = room_width - obj_depth
    y = (room_depth - obj_width) / 2
    return x, y


def circle_polygon(
    cx: float,
    cy: float,
    radius: float,
    segments: int = 48,
):
    """Return polygon approximation of a circle."""
    angles = np.linspace(
        0,
        2 * np.pi,
        segments,
        endpoint=False,
    )

    return [
        (
            cx + radius * np.cos(a),
            cy + radius * np.sin(a),
        )
        for a in angles
    ]


# =========================================================
# FURNITURE
# =========================================================

COUCH_WIDTH = 2.40
COUCH_DEPTH = 0.97
COUCH_HEIGHT = 0.90

couch_x, couch_y = centered_against_wall(
    SALA_WIDTH,
    SALA_DEPTH,
    COUCH_WIDTH,
    COUCH_DEPTH,
)

couch_coords = rectangle(
    couch_x,
    couch_y,
    COUCH_DEPTH,
    COUCH_WIDTH,
)


# =========================================================
# SCENE OBJECTS
# =========================================================

objects = [
    BoxObject(
        name="SALA",
        coords=rectangle(
            0,
            0,
            SALA_WIDTH,
            SALA_DEPTH,
        ),
        height=3.3,
        color="skyblue",
    ),
    BoxObject(
        name="STAGE",
        coords=rectangle(
            -STAGE_WIDTH,
            0,
            STAGE_WIDTH,
            STAGE_DEPTH,
        ),
        height=STAGE_HEIGHT,
        color="saddlebrown",
    ),
    BoxObject(
        name="COUCH",
        coords=couch_coords,
        height=COUCH_HEIGHT,
        color="saddlebrown",
    ),

    # Ceiling-mounted disks
    BoxObject(
        name="TML",
        coords=circle_polygon(
            TMp_x,
            TMLp_y,
            radius=0.10,  # 20 cm diameter
        ),
        height=0.001,      # 1 mm thick
        base_z=3.299,      # flush with 3.3 m ceiling
        color="gold",
    ),
    BoxObject(
        name="TMR",
        coords=circle_polygon(
            TMp_x,
            TMRp_y,
            radius=0.10,
        ),
        height=0.001,
        base_z=3.299,
        color="gold",
    ),
]


# =========================================================
# MESH GENERATION
# =========================================================

def extrude_polygon(
    coords,
    height,
    color,
    name,
    base_z=0.0,
    opacity=0.7,
):
    coords = np.asarray(coords)
    n = len(coords)

    bottom = np.column_stack(
        [coords, np.full(n, base_z)]
    )

    top = np.column_stack(
        [coords, np.full(n, base_z + height)]
    )

    vertices = np.vstack([bottom, top])

    x, y, z = vertices.T

    i, j, k = [], [], []
    offset = n

    # Bottom face
    for t in range(1, n - 1):
        i.append(0)
        j.append(t)
        k.append(t + 1)

    # Top face
    for t in range(1, n - 1):
        i.append(offset)
        j.append(offset + t + 1)
        k.append(offset + t)

    # Side walls
    for s in range(n):
        nxt = (s + 1) % n

        i.extend([s, s])
        j.extend([nxt, offset + nxt])
        k.extend([offset + nxt, offset + s])

    return go.Mesh3d(
        x=x,
        y=y,
        z=z,
        i=i,
        j=j,
        k=k,
        color=color,
        opacity=opacity,
        flatshading=True,
        name=name,
    )


# =========================================================
# FIGURE
# =========================================================

fig = go.Figure()

label_x = []
label_y = []
label_z = []
label_text = []

for obj in objects:
    fig.add_trace(
        extrude_polygon(
            obj.coords,
            obj.height,
            obj.color,
            obj.name,
            base_z=obj.base_z,
        )
    )

    center = np.mean(obj.coords, axis=0)

    label_x.append(center[0])
    label_y.append(center[1])
    label_z.append(
        obj.base_z + obj.height + 0.15
    )
    label_text.append(obj.name)


fig.add_trace(
    go.Scatter3d(
        x=label_x,
        y=label_y,
        z=label_z,
        mode="text",
        text=label_text,
        textposition="middle center",
        textfont=dict(
            size=16,
            color="black",
        ),
        showlegend=False,
    )
)

fig.update_layout(
    title="Interactive 3D Polygon Scene",
    width=1100,
    height=850,
    scene=dict(
        xaxis_title="X (m)",
        yaxis_title="Y (m)",
        zaxis_title="Z (m)",
        aspectmode="data",
        camera=dict(
            eye=dict(
                x=1.8,
                y=-2.0,
                z=1.4,
            )
        ),
    ),
)

# =========================================================
# CEILING REFERENCE LINES
# =========================================================

distance_from_wall = round(SALA_WIDTH - TM_p_x,2)

fig.add_trace(
    go.Scatter3d(
        x=[TMp_x, SALA_WIDTH],
        y=[TMLp_y, TMLp_y],
        z=[3.3, 3.3],
        mode="lines",
        line=dict(
            color="red",
            width=6,
        ),
        name=f"Distance from wall = {distance_from_wall}",
    )
)

fig.add_trace(
    go.Scatter3d(
        x=[TMp_x, SALA_WIDTH],
        y=[TMRp_y, TMRp_y],
        z=[3.3, 3.3],
        mode="lines",
        line=dict(
            color="red",
            width=6,
        ),
        name=f"Distance from wall = {distance_from_wall}",
    )
)

# Line 3
fig.add_trace(
    go.Scatter3d(
        x=[SALA_WIDTH, SALA_WIDTH],
        y=[TMLp_y, SALA_DEPTH / 2],
        z=[3.3, 3.3],
        mode="lines",
        line=dict(
            color="green",
            width=6,
        ),
        name=f"Distance from center = {side_shift}",
    )
)

# Line 4
fig.add_trace(
    go.Scatter3d(
        x=[SALA_WIDTH, SALA_WIDTH],
        y=[TMRp_y, SALA_DEPTH / 2],
        z=[3.3, 3.3],
        mode="lines",
        line=dict(
            color="green",
            width=6,
        ),
        name=f"Distance from center = {side_shift}",
    )
)

fig.add_trace(
    go.Scatter3d(
        x=[
            TMp_x, SALA_WIDTH,      # Line 1
            TMp_x, SALA_WIDTH,      # Line 2
            SALA_WIDTH, SALA_WIDTH, # Line 3
            SALA_WIDTH, SALA_WIDTH, # Line 4
        ],
        y=[
            TMLp_y, TMLp_y,
            TMRp_y, TMRp_y,
            TMLp_y, SALA_DEPTH / 2,
            TMRp_y, SALA_DEPTH / 2,
        ],
        z=[
            3.3, 3.3,
            3.3, 3.3,
            3.3, 3.3,
            3.3, 3.3,
        ],
        mode="markers",
        marker=dict(
            size=6,
            color="black",
        ),
        name="Line Endpoints",
    )
)

fig.show(renderer="browser")

```
